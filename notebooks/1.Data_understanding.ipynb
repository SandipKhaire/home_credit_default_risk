```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# Training Data
train_categories = np.array(['C1', 'C2', 'C3', 'C1', 'C2']).reshape(-1, 1)
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)  # Ignore unknown categories
encoder.fit(train_categories)  # Fit on training data

# Prediction Data
new_data = np.array(['C2']).reshape(-1, 1)
new_encoded = encoder.transform(new_data)  # Ensures all expected columns exist

print("\nOne-Hot Encoded Prediction Data:")
print(new_encoded)
print("Feature Names:", encoder.get_feature_names_out())
```
